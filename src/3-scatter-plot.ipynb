### Exercise 3: Scatter Plot with Annotations

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create the scatter plot
plt.scatter(x, y, color='blue', marker='o')

# Annotate each point with its (x, y) value
for i, j in zip(x, y):
    plt.text(i, j, f'({i}, {j})', fontsize=10, ha='right')

# Customize the plot
plt.title("Scatter Plot with Annotations", fontsize=14)  # Set the title
plt.xlabel("X-axis", fontsize=12)  # Label for the X-axis
plt.ylabel("Y-axis", fontsize=12)  # Label for the Y-axis
plt.grid(True, linestyle='--', alpha=0.7)  # Add gridlines for better visualization

# Display the plot
plt.show()
```
