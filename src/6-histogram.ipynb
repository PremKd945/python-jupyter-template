### Exercise 6: Histogram Customization

```python
import matplotlib.pyplot as plt

# Data
data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5]

# Create the histogram
plt.hist(data, bins=5, color='skyblue', edgecolor='black')

# Customize the plot
plt.title("Customized Histogram", fontsize=14)  # Add title
plt.xlabel("Values", fontsize=12)              # Label for the X-axis
plt.ylabel("Frequency", fontsize=12)           # Label for the Y-axis
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Add gridlines for the Y-axis

# Display the plot
plt.show()
```
