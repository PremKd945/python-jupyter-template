### Exercise 1: Line Plot Customization

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate x values ranging from 0 to 10
x = np.linspace(0, 10, 100)  # 100 points for a smooth curve

# Compute y values as the square of x
y = x**2

# Create the plot
plt.plot(x, y, label="y = x²", color="blue", linestyle="--", linewidth=2)

# Customize the plot
plt.title("Square Function: y = x²", fontsize=14)  # Add title
plt.xlabel("X-axis (Input Values)", fontsize=12)   # Add label for X-axis
plt.ylabel("Y-axis (Squared Values)", fontsize=12) # Add label for Y-axis
plt.grid(True)                                    # Enable grid
plt.legend()                                      # Add legend for better clarity

# Display the plot
plt.show()
```
