### Exercise 2: Bar Plot with Colors

```python
import matplotlib.pyplot as plt

# Data
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['red', 'blue', 'green', 'orange']  # Different colors for each bar

# Create the bar plot
plt.bar(categories, values, color=colors)

# Customize the plot
plt.title("Category Values", fontsize=14)  # Add title to the plot
plt.xlabel("Categories", fontsize=12)      # Label for the X-axis
plt.ylabel("Values", fontsize=12)          # Label for the Y-axis
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Add gridlines for the Y-axis

# Display the plot
plt.show()
```
