### Exercise 5: Subplot Layout

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the plots
x = np.linspace(0, 10, 100)
y = x**2
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]
pie_labels = ['Python', 'Java', 'C++', 'JavaScript']
pie_sizes = [40, 25, 20, 15]

# Create the 2x2 subplot layout
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Line chart (1st subplot)
axs[0, 0].plot(x, y, color='blue', label='y = x')
axs[0, 0].set_title("Line Chart")
axs[0, 0].set_xlabel("X-axis")
axs[0, 0].set_ylabel("Y-axis")
axs[0, 0].legend()
axs[0, 0].grid(True)

# Bar chart (2nd subplot)
axs[0, 1].bar(categories, values, color=['red', 'blue', 'green', 'orange'])
axs[0, 1].set_title("Bar Chart")
axs[0, 1].set_xlabel("Categories")
axs[0, 1].set_ylabel("Values")
axs[0, 1].grid(axis='y', linestyle='--', alpha=0.7)

# Scatter plot (3rd subplot)
axs[1, 0].scatter(x_scatter, y_scatter, color='purple', marker='o')
for i, j in zip(x_scatter, y_scatter):
    axs[1, 0].text(i, j, f'({i}, {j})', fontsize=10, ha='right')
axs[1, 0].set_title("Scatter Plot")
axs[1, 0].set_xlabel("X-axis")
axs[1, 0].set_ylabel("Y-axis")
axs[1, 0].grid(True, linestyle='--', alpha=0.7)

# Pie chart (4th subplot)
axs[1, 1].pie(pie_sizes, labels=pie_labels, autopct='%1.1f%%', startangle=90, colors=['gold', 'skyblue', 'lightgreen', 'coral'])
axs[1, 1].set_title("Pie Chart")

# Adjust layout
plt.tight_layout()

# Display the plots
plt.show()
```
